```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data={'Transaction type':['Money in' ,'money out'],
"SEND MONEY":('0.00','66,309.00'),
"RECEIVED MONEY":(' 79,941.92','0.00' ),
"AGENT DEPOSIT": ('3,700.00' ,'0.00'),
"AGENT WITHDRAWAL":('0.00', '5,516.00'),
"LIPA NA M-PESA (PAYBILL)": ('0.00','29,608.00'),
"LIPA NA M-PESA (BUY GOODS)": ('0.00' ,'15,867.00'),
"OTHERS": ('33,627.46',' 9,120.74'),
"TOTAL": ('122,201.83',' 126,420.74')}
df=pd.DataFrame(data)
print(df)
# Display basic information about the dataframe
print(df.info())
# Describe the numerical columns
print(df.describe())
```

```python
# For the pie chart, we will focus on the 'TOTAL' values for simplicity
money_in = float(df['TOTAL'][0].replace(',', ''))
money_out = float(df['TOTAL'][1].replace(',', ''))

# Data for the pie chart
labels = ['Money In', 'Money Out']
sizes = [money_in, money_out]

# Plotting the pie chart
plt.figure(figsize=(4,4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Total Money In vs Money Out')
plt.axis('equal')  
# Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```

```python

# Extract and sum relevant values for 'Money In' and 'Money Out'
money_in_labels = ['RECEIVED MONEY', 'AGENT DEPOSIT']
money_out_labels = ['SEND MONEY', 'AGENT WITHDRAWAL', 'LIPA NA M-PESA (PAYBILL)', 'LIPA NA M-PESA (BUY GOODS)']

money_in_values = [float(df[label][0].replace(',', '')) for label in money_in_labels]
money_out_values = [float(df[label][1].replace(',', '')) for label in money_out_labels]

# Data for the pie chart
labels = money_in_labels + money_out_labels
sizes = money_in_values + money_out_values

# Plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Transaction Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```
